```python
# Structure of schedule 
slots_per_day = 6
class_days = 5
total_cohorts = 8

# There will be five matrices, one for each class day (with 6 rows and 8 columns).
# Within each of the matrices, there will be lists of slots (6 per day).
# Each column will represent a different cohort (8 in total).

week_schedule = list()
for i in range(class_days):
    day = list()
    for j in range(slots_per_day):
        slots=list()
        for k in range(total_cohorts):
            cohort=list()
            slots.append(cohort)
        day.append(slots)
    week_schedule.append(day)
```

```python
class Teacher(object):
    
    def __init__(self,name,day_available,classes_given,cohorts):
        self.name = name # name of teacher (string)
        self.day_available = day_available # list of days in which teacher can come to the school
        self.classes_given = classes_given # total amount of classes given
        self.cohorts = cohorts # dictionary with cohorts taught and amount of classes per cohort
        
        # amount of classes to be scheduled
        self.class_count = classes_given
        
    def calculate_score(self, class_count, day_available):
        """
        OUTPUT: The availability score.
        """
        if len(self.day_available) == 0:
            return 0
        else:
            return self.class_count / len(self.day_available)
    
    def update_class_count(self,class_count):
        self.class_count = self.class_count-1
```

```python
def heapify_scheduler(priority, i):
    """
    NOTE: This implementation was adapted from the Assignment 2.
    
    GOAL: Implement the heapify function based on the duration of the task.
        Scheduling criteria: lengthier tasks first.
        
    INPUT: 
    - Priority list with priority task objects.
    - Index i of the element currently being compared.
    
    OUTPUT: semi-heap given that index
    """    
    def left(i):
        return 2*i+1
    
    def right(i):
        return 2*i+2
    
    # creating left and right children
    l=left(i)
    r=right(i)
    largest=i
    
    # evaluating left    
    if l < len(priority):
        if priority[l][0] > priority[i][0]:
            largest = l

    # evaluating right   
    if r < (len(priority)):
        if priority[r][0] > priority[largest][0]:          
            largest = r

        # switching positions
    if largest != i:
        priority[i], priority[largest] = priority[largest],priority[i]

        ## HEAPIFYING RECURSIVELY
        heapify_scheduler(priority,largest)
    
    return priority
```

```python
def determining_start(prof_list,cohort_list,day_of_week):
    """
    INPUT:
    - List with all the cohorts name (keys in the dictionary of cohorts)).
    - List of teacher objects.
    
    OUTPUT: Ordered list of cohorts to go through. 
    """
    final_order = list() 
    for cohort in cohort_list:
    
        # select teachers available to teach the cohort
        teacher_cohort = list()
        for prof in prof_list:
            if prof.cohorts[cohort] > 0 and day_of_week in prof.day_available:
                teacher_cohort.append(prof)
                
        # compute availability score
        av_sum = 0
        for item in teacher_cohort:
            av_score = item.calculate_score(item.class_count, item.day_available)
            av_sum +=av_score
            
        final_order.append([av_sum,cohort])
        
    # put final_order in order based on their av_sum
    for j in range(int(len(final_order)/2),-1,-1): 
        heapify_scheduler(final_order, j)
    
    return final_order
```

```python
def create_heap(prof_list,day_of_week,cohort):
    """
    INPUT:
    - prof_list: list of teachers objects.
    - day of the week we are refering to (between 1 and 5)
    - cohort we are refering to.
    
    OUTPUT: Given a given day and cohort, create a max heap with the av_scores of the teachers available.
    """
    # select teachers available to teach the cohort
    teacher_cohort = list()
    for prof in prof_list:
        if prof.cohorts[cohort] > 0 and len(prof.day_available) > 0 and prof.class_count > 0 and day_of_week in prof.day_available:
            teacher_cohort.append(prof)
    
    # compute the av_scores of every teacher
    heap=list()
    for item in teacher_cohort:
        av_score = item.calculate_score(item.class_count, item.day_available)
        heap.append([av_score,item])
    
    # create the heap
    for j in range(int(len(heap)/2),-1,-1): 
        heapify_scheduler(heap,j)
        
    return heap
```

```python
def assigning_root(heap_cohort, day_of_week, cohort, slot, schedule):
    """
    INPUT:
    - heap_cohort: Heap with teachers objects and their availability scores for given day and cohort.
    - Day of the week (1-5, for five days in the week)
    - Cohort number (0:6A,1:6B,2:7A,3:7B, 4:8A,5:8B,6:9A,7:9B)
    - Slot number (0-5, for six classes)
    - Schedule: list of lists on which we will include the schedules.
    
    OUTPUT: 
    - Edited heap with new av_score and order.
    - Slot filled for the given cohort and given day.
    """

    # list with all the teachers who are scheduled to teach 
    slot_all_cohorts=schedule[day_of_week-1][slot]
    
    # specific slot in which we are looking to include a teacher on
    slot_to_fill = schedule[day_of_week-1][slot][cohort]
    
    # edge case: if the heap is empty from the start
    if len(heap_cohort) == 0:
        #print(f"No teacher available for {cohort} on  day {day_of_week}")
        slot_to_fill = 'TBD' 
        return slot_to_fill
    
    # potential item to be included
    ## which is a list whose first element (index 0) is the AV score and the second (index 1) the teacher object
    else:    
        index=0
        item = heap_cohort[index]
    
    # verify if the root is already included in the slot we are inserting on
    while item[1].name in slot_all_cohorts and (index+1) < len(heap_cohort):
        # if it is, get the next one
        index+=1
        item=heap_cohort[index]
        #print("Already exists! Must change.")
        
    # if all of the teachers available are already included in the slot
    if index == len(heap_cohort):
        #print(f"No teacher available for {cohort} on  day {day_of_week}")
        slot_to_fill = 'TBD' 
        return slot_to_fill
    
    # if not, insert the teacher in the slot
    else: 
        schedule[day_of_week-1][slot][cohort] = item[1].name
        
    # printing test 
    # print(schedule[day_of_week-1][slot])

    # re-compute and class_count for root
    item[1].update_class_count(item[1].class_count)
    
    # re-compute the av_score for root
    item_av_score = item[1].calculate_score(item[1].class_count, item[1].day_available)
    
    # edit the item av_score in the heap
    ## find item in the heap
    if item_av_score > 0:
        heap_cohort[index][0] = item_av_score
        
    # remove if av_score is zero
    else:
        heap_cohort.remove(item) 
    
    # heapify again with new 
    for j in range(int(len(heap_cohort)/2),-1,-1): 
        heapify_scheduler(heap_cohort,j)
        
    #print(f"new heap:")
    #for item in heap_cohort:
    #    print(f"{item[0]} for teacher {item[1].name}")
    
    return slot_to_fill
```

```python
def translate_cohort_number(cohort_name):
    """
    INPUT:
    - cohort name we want to translate to number.
    OUTPUT: cohort number
    WARNING - this is specific to this specific solution.
    """
    if cohort_name=='6A':
            cohort_number =0
    elif cohort_name=='6B':
            cohort_number =1
    elif cohort_name=='7A':
        cohort_number =2
    elif cohort_name=='7B':
        cohort_number =3
    elif cohort_name=='8A':
        cohort_number =4
    elif cohort_name=='8B':
        cohort_number =5
    elif cohort_name=='9A':
        cohort_number =6
    elif cohort_name=='9B':
        cohort_number =7
    else:
        print("This cohort is not on the list.")
        cohort_number=None
    return cohort_number
```

```python
def implementing_day(prof_list, cohort_list, day_of_week, slot_amount, schedule):
    """
    INPUT: 
    - List of teacher objects
    - List of cohort names
    - Day of week (1-5)
    - Amount of slots
    - Schedule - list of lists where we should input the schedule
    OUTPUT: The schedule for all cohorts for a given day
    """
    # create the ordered cohorts list
    order_cohorts = determining_start(prof_list,cohort_list,day_of_week)
    
    ## create a heap for the professors available on that day
    for cohort in order_cohorts:
        #print(" ----- ")
        #print(f"Currently scheduling: {cohort}")
        #print("")
        heap_cohort = create_heap(prof_list,day_of_week,cohort[1])
        
    ## assign root to each of the slots 
        cohort_number = translate_cohort_number(cohort[1])
        for i in range(slot_amount):
            assigning_root(heap_cohort, day_of_week, cohort_number, i, schedule)
            
    ## exclude one day count for all teachers scheduled
    for prof in profs:
        if day_of_week in prof.day_available:
            prof.day_available.remove(day_of_week)
    
    return schedule
```

```python
def implementing_week(prof_list, cohort_list, slot_amount, schedule):
    """
    INPUT:
    - List of teacher objects
    - List of cohort names
    - Amount of slots in a day
    - Schedule - list of lists where we should input the schedule
    OUTPUT: The schedule for all cohorts for the entire week.
    """
    for i in range(1,6):
        implementing_day(prof_list, cohort_list, i, slot_amount, schedule)    
    return schedule
```
